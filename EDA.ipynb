# **8-10. X & Y & Z - EDA**

Below we are going to create various graph plots to view the relationship between our length, width, and depth of the diamonds. 

We have created the subplots all of the three variables to be able to view the relationship with the depth for both the length and the width. 

```python
import numpy as np
import matplotlib.pyplot as plt
x1 = df['z']
x2 = df['z']

y1 = df['x']
y2 = df['y']

fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('Diamond Legth, Width, and Depth')

ax1.plot(x1, y1, 'o-')
ax1.set_ylabel('Length in mm')

ax2.plot(x2, y2, '.-')
ax2.set_xlabel('Depth')
ax2.set_ylabel('Width in mm')

plt.show()
```

This first visual above provides a visual of the concentration of diamonds that have a depth between 0-5 and generally have either a width of 0-20 or a length of 0-10. Though it is not completely apparent there is a linear relationship between the length with the depth and the width with the depth. There are some outliers that appear to be data quality errors with a width of 60 or a length of 30. 

```python
import numpy as np
import matplotlib.pyplot as plt
x1 = df['y']
x2 = df['y']

y1 = df['x']
y2 = df['z']

fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('Diamond Legth, Width, and Depth')

ax1.plot(x1, y1, 'o-')
ax1.set_ylabel('Length in mm')

ax2.plot(x2, y2, '.-')
ax2.set_xlabel('Width')
ax2.set_ylabel('Depth in mm')

plt.show()
```

**1. What proportion of diamonds are between .30 and 1.08 carats?**

We will calculate a dataframe that contains diamonds within this range. Then we will calculate the proportion of that dataframe to the total dataset.

```python
betweencarats = df[(df['carat'] >=.3) & (df['carat'] <=1.08)]
countbtween = betweencarats['carat'].count()
total = df['carat'].count()
proportion = countbtween/total
percentage = "{:.0%}". format(proportion)
print('The proportion of diamonds between .30 and 1.08 carats to the total diamonds is ' + percentage)
```

**2. How many of the diamonds have equal x and y dimensions?**

We will create a dataframe that only includes diamonds that have equal lengths and widths (x and y). Then we will do a count on that dataframe. 

```python
equalxy = df[(df['x'] == df['y'])]
equalxy['carat'].count()
```

**3. How many of the diamonds have a carat value that is less than the mean carat value?**

First we must identify the mean carat value. Once we create a dataframe with that value we can create an additional dataframe that can filter the values based off of the mean. Lastly we do a count of our new filtered dataframe to get the result. 

```python
#identify the mean and then create a dataframe with all diamonds less
#then count the records in that dataframe
mean = df['carat'].mean()
less_mean = df[df['carat'] < mean]
less_mean['carat'].count()
```

**5. Which diamond has the highest price per carat? What is its value?**


We used numpy for this question and calculated the price per carat value and stored it in a new dataframe. After this we needed to identify the max which we did for our new "pricepercarat" variable.

```python
import numpy as np
#calculate the price per carat and then get the max from that output
df['pricepercarat'] = df['price']/df['carat']
df['pricepercarat'].max()

```

We used the argmax function to derive the exact diamond in our data set that has the highest price per diamond. To identify which diamond it was we needed that function to know the row number/index. 

```python
#use numpys argmax function to return the row number for the max
df['pricepercarat'].argmax()
```
